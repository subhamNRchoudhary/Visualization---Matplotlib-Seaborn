```python
import seaborn as sns
import pandas as pd
import numpy as np


iris = sns.load_dataset('iris')


np.random.seed(0)
n = 1000
synthetic_data = pd.DataFrame({
    'A': np.random.normal(0, 1, n),
    'B': np.random.normal(0, 2, n),
    'C': np.random.normal(5, 1, n),
    'D': np.random.uniform(0, 10, n),
    'Category': np.random.choice(['Group 1', 'Group 2', 'Group 3'], n)
})

#  first few rows of each dataset
print(iris.head(5))
print(synthetic_data.head())
```

# Matplotlib Visualizations

```python
import matplotlib.pyplot as plt

# Line plot
plt.figure(figsize=(10, 7))
plt.plot(synthetic_data['A'], label='A')
plt.plot(synthetic_data['B'], label='B')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Line Plot')
plt.legend()
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(synthetic_data['A'], synthetic_data['B'], c=synthetic_data['D'], cmap='viridis')
plt.colorbar(label='D')
plt.xlabel('A')
plt.ylabel('B')
plt.title('Scatter Plot')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
plt.hist(synthetic_data['A'], bins=30, alpha=0.5, label='A')
plt.hist(synthetic_data['B'], bins=30, alpha=0.5, label='B')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.legend()
plt.show()
```

```python
# Bar plot
category_counts = synthetic_data['Category'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(category_counts.index, category_counts.values)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Bar Plot')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(10, 6))
plt.boxplot([synthetic_data['A'], synthetic_data['B'], synthetic_data['C']], labels=['A', 'B', 'C'])
plt.ylabel('Value')
plt.title('Box Plot')
plt.show()
```

```python
# Pie chart
category_counts = synthetic_data['Category'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.show()
```

```python
# Area plot
plt.figure(figsize=(10, 6))
plt.fill_between(range(n), synthetic_data['A'], color="skyblue", alpha=0.4)
plt.fill_between(range(n), synthetic_data['B'], color="sandybrown", alpha=0.5)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Area Plot')
plt.show()
```

```python
# Stacked bar plot
stacked_data = synthetic_data.groupby('Category').sum()
stacked_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Stacked Bar Plot')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# 3D scatter plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(synthetic_data['A'], synthetic_data['B'], synthetic_data['C'], c=synthetic_data['D'], cmap='viridis')
ax.set_xlabel('A')
ax.set_ylabel('B')
ax.set_zlabel('C')
plt.title('3D Scatter Plot')
plt.show()
```

```python
# Hexbin plot
plt.figure(figsize=(10, 6))
plt.hexbin(synthetic_data['A'], synthetic_data['B'], gridsize=30, cmap='Blues')
plt.colorbar(label='count')
plt.xlabel('A')
plt.ylabel('B')
plt.title('Hexbin Plot')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Generate synthetic data
np.random.seed(42)
n = 100
synthetic_data = pd.DataFrame({
    'A': np.random.rand(n),
    'B': np.random.rand(n),
    'C': np.random.rand(n),
    'D': np.random.rand(n),
    'Category': np.random.choice(['Category 1', 'Category 2', 'Category 3'], n)
})
iris = sns.load_dataset("iris")
```

```python
# Matplotlib Visualizations

# 1. Line Plot
plt.figure(figsize=(10, 6))
plt.plot(synthetic_data['A'], color='navy')
plt.title('Line Plot')
plt.show()
```

```python
# 2. Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(synthetic_data['A'], synthetic_data['B'], color='Green')
plt.xlabel('A')
plt.ylabel('B')
plt.title('Scatter Plot')
plt.show()
```

```python
# 3. Histogram
plt.figure(figsize=(10, 6))
plt.hist(synthetic_data['A'], bins=20, color='turquoise')
plt.xlabel('A')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# 4. Bar Plot
category_counts = synthetic_data['Category'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(category_counts.index, category_counts.values, color='teal')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Bar Plot')
plt.show()
```

```python
# 5. Box Plot
plt.figure(figsize=(10, 6))
plt.boxplot([synthetic_data['A'], synthetic_data['B'], synthetic_data['C']], patch_artist=True,
            boxprops=dict(facecolor='olive'))
plt.xlabel('Feature')
plt.ylabel('Value')
plt.title('Box Plot')
plt.show()
```

```python
# 6. Pie Chart
category_counts = synthetic_data['Category'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140, colors=['maroon', 'chocolate', 'orchid'])
plt.title('Pie Chart')
plt.show()
```

```python
# 7. Area Plot
plt.figure(figsize=(10, 6))
plt.fill_between(range(n), synthetic_data['A'], color="skyblue", alpha=0.4)
plt.fill_between(range(n), synthetic_data['B'], color="sandybrown", alpha=0.5)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Area Plot')
plt.show()
```

```python
# 8. Stacked Bar Plot
stacked_data = synthetic_data.groupby('Category').sum()
stacked_data.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#FF5733', '#33FF57', '#3357FF'])
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Stacked Bar Plot')
plt.show()
```

```python
# 9. 3D Scatter Plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(synthetic_data['A'], synthetic_data['B'], synthetic_data['C'], c=synthetic_data['D'], cmap='viridis')
ax.set_xlabel('A')
ax.set_ylabel('B')
ax.set_zlabel('C')
plt.title('3D Scatter Plot')
plt.show()
```

```python
# 10. Hexbin Plot
plt.figure(figsize=(10, 6))
plt.hexbin(synthetic_data['A'], synthetic_data['B'], gridsize=30, cmap='Blues')
plt.colorbar(label='count')
plt.xlabel('A')
plt.ylabel('B')
plt.title('Hexbin Plot')
plt.show()
```

```python
# Seaborn Visualizations

# 1. Pairplot
sns.pairplot(synthetic_data, hue='Category', palette='Set2')
plt.show()
```

```python
# 2. Heatmap
corr = synthetic_data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

```python
# 3. Violin Plot
sns.violinplot(x='Category', y='A', data=synthetic_data, palette='Pastel1')
plt.title('Violin Plot')
plt.show()
```

```python
# 6. Point Plot
sns.pointplot(x='Category', y='A', data=synthetic_data, palette='Dark2')
plt.title('Point Plot')
plt.show()
```

```python
# 7. Strip Plot
sns.stripplot(x='Category', y='A', data=synthetic_data, palette='Accent')
plt.title('Strip Plot')
plt.show()
```
